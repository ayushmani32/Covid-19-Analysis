#  Prediction Using support vector

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
sc_y= StandardScaler()
Sx= sc_X.fit_transform(x)
Sy= sc_y.fit_transform(y)
```

```python
Sx
```

```python
from sklearn.svm import SVR
reg= SVR(kernel='rbf')
reg.fit(Sx,Sy.ravel())
```

```python
plt.scatter(Sx,Sy)
plt.plot(Sx,reg.predict(Sx),color="k",linewidth=5)
plt.show()
```

```python
reg.score(Sx,sy)*100
```
